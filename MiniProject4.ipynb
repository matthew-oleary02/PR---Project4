```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('./Mall_customers.csv', skiprows=1, header = None, names = ['Gender','Age','AnnualIncome','SpendingScore'])
```

```python
data['Gender'].replace('Female', 0,inplace=True)
data['Gender'].replace('Male', 1,inplace=True)
data.head()
```

```python
spendscore = data[['SpendingScore']]
samples = data[['Gender', 'Age', 'AnnualIncome']]
samples.head()
```

```python
ks = range(1, 10)
inertias = []

for k in ks:
   # Create a KMeans instance with k clusters: model
   model = KMeans(n_clusters = k)

   # Fit model to samples
   model.fit(samples)

   # Append the inertia to the list of inertias
   inertias.append(model.inertia_)

# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
model = KMeans(n_clusters = 5)
model.fit(samples)

labels = model.predict(samples)
```

```python
data["labels"] = labels
#data.head(50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import pandas as pd

# Assuming 'data' is your existing DataFrame with 'AnnualIncome' and 'SpendingScore'
# Replace this with actual data loading if necessary
# Example: data = pd.read_csv('your_data.csv')

# Check if the label columns exist in the DataFrame
print(data.columns)

# Ensure each label column exists and is properly populated
if 'labels_3' not in data.columns or data['labels_3'].isnull().all():
    print("labels_3 is missing or not populated")
    
if 'labels_4' not in data.columns or data['labels_4'].isnull().all():
    print("labels_4 is missing or not populated")
    
if 'labels_5' not in data.columns or data['labels_5'].isnull().all():
    print("labels_5 is missing or not populated")
    
if 'labels_6' not in data.columns or data['labels_6'].isnull().all():
    print("labels_6 is missing or not populated")

# If labels are missing, generate them using KMeans
X = data[['AnnualIncome', 'SpendingScore']]  # Columns for clustering

# Generate labels for 3, 4, 5, and 6 clusters
kmeans_3 = KMeans(n_clusters=3, random_state=42).fit(X)
kmeans_4 = KMeans(n_clusters=4, random_state=42).fit(X)
kmeans_5 = KMeans(n_clusters=5, random_state=42).fit(X)
kmeans_6 = KMeans(n_clusters=6, random_state=42).fit(X)

# Assign the labels to the DataFrame
data['labels_3'] = kmeans_3.labels_
data['labels_4'] = kmeans_4.labels_
data['labels_5'] = kmeans_5.labels_
data['labels_6'] = kmeans_6.labels_

# Plotting clusters
label_columns = ['labels_3', 'labels_4', 'labels_5', 'labels_6']
titles = ['3 Clusters', '4 Clusters', '5 Clusters', '6 Clusters']
palette_list = [['green', 'orange', 'brown'], 
                ['green', 'orange', 'brown', 'dodgerblue'], 
                ['green', 'orange', 'brown', 'dodgerblue', 'red'], 
                ['green', 'orange', 'brown', 'dodgerblue', 'red', 'purple']]

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

# Loop over the axes and plot the clusters
for idx, ax in enumerate(axes.flat):
    sns.scatterplot(x='AnnualIncome', y='SpendingScore', hue=label_columns[idx],
                    palette=palette_list[idx], legend='full', data=data, s=60, ax=ax)
    ax.set_title(f'Spending Score vs Annual Income: {titles[idx]}')
    ax.set_xlabel('Annual Income')
    ax.set_ylabel('Spending Score')
    ax.legend(title=f'{len(palette_list[idx])} Clusters')

plt.suptitle('Comparison of Clusters (3, 4, 5, 6)', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Define the list of clusters and angles
configurations = [
    {'clusters': 4, 'elev': 30, 'azim': 60, 'title': '4 Clusters (30, 60)'},
    {'clusters': 5, 'elev': 30, 'azim': 60, 'title': '5 Clusters (30, 60)'},
    {'clusters': 4, 'elev': 30, 'azim': 120, 'title': '4 Clusters (30, 120)'},
    {'clusters': 4, 'elev': 45, 'azim': 90, 'title': '4 Clusters (45, 90)'}
]

# Loop through each configuration and generate a scatter plot
fig = plt.figure(figsize=(20, 12))

for i, config in enumerate(configurations):
    ax = fig.add_subplot(2, 2, i + 1, projection='3d')
    
    # Use the cluster label from the configuration (modify data['labels'] accordingly)
    clusters = data[f'labels_{config["clusters"]}'] if f'labels_{config["clusters"]}' in data.columns else data['labels']
    
    # Create a scatter plot
    scatter = ax.scatter(data['AnnualIncome'], data['SpendingScore'], data['Age'],
                         c=clusters, cmap='viridis', s=60)
    
    # Add axis labels
    ax.set_xlabel('Annual Income')
    ax.set_ylabel('Spending Score')
    ax.set_zlabel('Age')
    
    # Set title
    ax.set_title(config['title'])
    
    # Adjust view angles
    ax.view_init(elev=config['elev'], azim=config['azim'])
    
    # Add legend
    legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
    ax.add_artist(legend1)

# Display all the plots
plt.tight_layout()
plt.show()
```
